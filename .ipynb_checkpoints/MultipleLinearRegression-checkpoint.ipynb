## A venture capitaliest company has hire you as an ML enginere to create a model that can predict of Profit of startup company based on companies Spending pattern and Location

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('dataset\\50_Startups.csv')
data
```

```python
# Seperate your data as Features and Lable
features = data.iloc[:,[0,1,2]].values
features
```

```python
label = data.iloc[:,[4]].values
label
```

```python
# Rule when working Regression using sklearn
# SKLearn expects your features and lebel to be in 2D form
print(features.shape)
print(label.shape)
```

```python
# Create training and testing set
# Training set will be used to test our model (learning)
# Testing set will be used to test our model accuracy/genralization (testing)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features,label,test_size=0.2,random_state=10)
```

```python
#Now we will create our Model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train) # training my model & Creating intercept and coefficent
```

```python
#Check the generalization
print(lr.score(X_train, y_train)) #Known data
print(lr.score(X_test, y_test)) #Unknown data
```

```python
# From above we know will give best value for random state = 10 so we will change it
# Now deploy the model
import pickle
pickle.dump(lr,open('PorfitPredictor.model','wb'))
```

```python
#Equation 

#Porfit = b0 + b1*(R&D)+b2(Adm)+b4*(Marketing)+b5*(Newyk)+B6*(Florid)   ## b0=intercept and b1=coefficient
print("The equation of regression line is Profit = {} + {}*R&D".format(lr.intercept_,lr.coef_))
```
