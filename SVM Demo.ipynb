```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('iris.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
featues = data.iloc[:,1:5].values
featues
```

```python
label = data.iloc[:,5].values
label
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(featues,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=10)
```

```python
from sklearn.svm import SVC  #Support Vector Classifier & SVC is the support vector machine algorithm fot the multiclass 

model = SVC(kernel='rbf',gamma='auto') #Best practice is to set Gamma inbetween 0 to 1 & rbf=radial bais function
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
